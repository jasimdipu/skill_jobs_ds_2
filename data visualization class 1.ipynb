# Seaborn

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip',data=tips, hue='sex')
```

```python
sns.relplot(x='total_bill', y='tip',data=tips, hue="sex", style='smoker')
```

```python
sns.relplot(x='total_bill', y='tip',data=tips, hue="sex", style='smoker', size='size')
```

```python
sns.relplot(x='total_bill', y='tip',data=tips, hue="sex", style='smoker', size='size', col='time')
```

```python
sns.relplot(x='total_bill', y='tip',data=tips, hue="sex", style='smoker', size='size', col='day')
```

```python
# catplot
sns.catplot(x='day', y='tip', data=tips)
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='swarm')
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='swarm', hue='size')
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='box')
```

```python
sns.catplot(x='day', y='tip', data=tips, kind='box', hue='size')
```

```python
sns.pairplot(data=tips)
```
